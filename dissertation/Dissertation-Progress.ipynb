```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_context('poster')
```

```python
df = pd.read_csv('progress.tsv', sep='\t', index_col='Date')
```

```python
df.head()
```

```python
fig, ax1 = plt.subplots()
df.Pages.plot(ax=ax1, label='Total (Pages)')
df.Content.plot(ax=ax1, label='Content (Pages)')
ax1.hlines(120, 0, 100, linestyles=':', colors='g', label='Content Goal (Pages)')
plt.title('Progress')
ax1.set_ylim([0,150])
ax1.set_ylabel('Pages')
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
df.Words.plot(ax=ax2, label='Words Written', grid=False, c='k')
ax2.set_ylim([0,3500])
ax2.hlines(500, 0, 100, linestyles=':', colors='r', label="Daily Word Goal")
ax2.set_ylabel('Words (Per Day)')

ax2.legend(loc=(0.0126,0.66))

                
```
