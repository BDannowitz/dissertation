```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context('notebook')
sns.set_style('whitegrid')
```

```python
colors = sns.color_palette()
```

```python
df = pd.read_csv('PMT_Test2.tsv', sep='\t', index_col='Frequency (Hz)')
#df.dropna(inplace=True)
df = df.ix[1:]
df
```

```python
fig, ax = plt.subplots()
df.plot(y='Original Amplitude', logx=True, ax=ax,
        marker='o', ls='-', lw=1, color=colors[0], ms=5)
df.plot(y='v1 Amplitude', logx=True, ax=ax,
        marker='o', ls='-', lw=1, color=colors[1], ms=5)
ax.set_title('Signal Amplitude')
ax.grid(axis='x', which='both', lw=1, ls=':')
ax.set_ylabel('-V', rotation=0, labelpad=15)
ax.legend(numpoints=2, loc='best')
```

```python
fig, ax = plt.subplots()
df.plot(y='Original Bleeder Current', logx=True, ax=ax,
        marker='o', ls='-', lw=1, color=colors[0], ms=5)
df.plot(y='Original Signal Current', logx=True, ax=ax,
        marker='s', ls='--', lw=1, color=colors[0], ms=5)
df.plot(y='v1 Bleeder Current', logx=True, ax=ax,
        marker='o', ls='-', lw=1, color=colors[1], ms=5)
df.plot(y='v1 Signal Current', logx=True, ax=ax,
        marker='s', ls='--', lw=1, color=colors[1], ms=5)
ax.set_title('PMT Currents')
ax.grid(axis='x', which='both', lw=1, ls=':')
ax.set_ylabel('mA', rotation=0, labelpad=15)
ax.legend(numpoints=2, loc='best')
```

```python
df2 = pd.read_csv('PMT_Test-Half Resistance.tsv', sep='\t', index_col='Frequency (Hz)')
df2.head()
```

```python
fig, ax = plt.subplots(1)
df2.plot(y='Original Amplitude', ax=ax,
         marker='o', ls='-', lw=1, color=colors[0], ms=5)
df2.plot(y='v1 Amplitude', logx=10, ax=ax,
         marker='o', ls='-', lw=1, color=colors[1], ms=5)
df2.plot(y='v2 Amplitude', logx=10, ax=ax,
         marker='o', ls='-', lw=1, color=colors[2], ms=5)
ax.grid(axis='x', which='both', lw=1, ls=':')
ax.set_title('Signal Amplitude')
ax.set_ylabel('-V', rotation=0, labelpad=15)
ax.legend(numpoints=2, loc='best')
```

```python
fig, ax = plt.subplots()
df2.plot(y='Original Bleeder Current', logx=True, ax=ax,
        marker='o', ls='-', lw=1, color=colors[0], ms=5)
df2.plot(y='Original Signal Current', logx=True, ax=ax,
        marker='s', ls='--', lw=1, color=colors[0], ms=5)
df2.plot(y='v1 Bleeder Current', logx=True, ax=ax,
        marker='o', ls='-', lw=1, color=colors[1], ms=5)
df2.plot(y='v1 Signal Current', logx=True, ax=ax,
        marker='s', ls='--', lw=1, color=colors[1], ms=5)
df2.plot(y='v2 Bleeder Current', logx=True, ax=ax,
        marker='o', ls='-', lw=1, color=colors[2], ms=5)
df2.plot(y='v2 Signal Current', logx=True, ax=ax,
        marker='s', ls='--', lw=1, color=colors[2], ms=5)
ax.set_title('PMT Currents')
ax.grid(axis='x', which='both', lw=1, ls=':')
ax.set_ylabel('mA', rotation=0, labelpad=15)
ax.legend(numpoints=2, loc=(0.02, 0.45))
```
